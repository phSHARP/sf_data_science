1) Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education_category(info: str) -> str:
    """Возвращает строку, обозначающую категорию образования, на основе
    полной информации об образовании.
    """
    if info.startswith('Высшее'):
        return 'высшее'
    elif info.startswith('Неоконченное'):
        return 'неоконченное высшее'
    elif info.startswith('Среднее специальное'):
        return 'среднее специальное'
    elif info.startswith('Среднее'):
        return 'среднее'
    else:
        return 'другое/отсутствует'


hh_data['Образование'] = hh_data['Образование и ВУЗ'] \
    .apply(get_education_category).astype('category')
display(hh_data['Образование'].unique())

secondary_education = len(hh_data[hh_data['Образование'] == 'среднее'])
display(f'Соискателей со средним образованием: {secondary_education}')

hh_data.drop('Образование и ВУЗ', axis='columns', inplace=True)
```

```python
def get_gender(info: str) -> str:
    """Возвращает пол из строки с информацией о соискателе."""
    if info.startswith('Мужчина'):
        return 'М'
    else:
        return 'Ж'


def get_age(info: str) -> int:
    """Возвращает возраст (число) из строки с информацией о соискателе."""
    age_info = info.split(',')[1]  # Полная информация о возрасте
    return int(age_info.split()[0])


hh_data['Пол'] = hh_data['Пол, возраст'].apply(get_gender).astype('category')
display(hh_data['Пол'].unique())
hh_data['Возраст'] = hh_data['Пол, возраст'].apply(get_age)
display(hh_data['Возраст'].describe())

female_percentage = len(hh_data[hh_data['Пол'] == 'Ж']) / len(hh_data) * 100
display(f'Процент женских резюме: {female_percentage:.2f}%')

display(f'Средний возраст соискателей: {hh_data["Возраст"].mean():.1f}')

hh_data.drop('Пол, возраст', axis='columns', inplace=True)
```

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience_months(info: str) -> int:
    """Возвращает количество месяцев, которые проработал соискатель."""
    
    if info is np.nan or info == 'Не указано':
        return np.nan
    
    words = info.split()
    years, months = int(words[2]), words[4]  # Наличие первого числа гарантируется
    if words[3].startswith('месяц'):
        years, months = 0, years
    else:
        months = int(months) if months.isnumeric() else 0
    
    return years*12 + months


hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience_months)
display(hh_data['Опыт работы (месяц)'].describe())

display(f'Медианный опыт работы: {hh_data["Опыт работы (месяц)"].median():.0f} месяцев')

hh_data.drop('Опыт работы', axis='columns', inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city_category(info: str) -> str:
    """Возвращает строку, обозначающую город или его категорию,
    из строки с информацией о городах, переездах, командировок.
    """
    
    category_cities = ['Москва', 'Санкт-Петербург']
    million_cities = [
        'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань',
        'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
        'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
    ]
    
    city = info.split(',')[0].rstrip()
    if city in category_cities:
        return city
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'


# Вспомогательная функция
def get_relocate_and_travel_will(info: str) -> tuple:
    """Возвращает желание соискателя к переезду и командировкам
    из строки с информацией о городах, переездах, командировок.
    """
    
    info = info.split(',')[1:]
    info = list(map(lambda s: s.strip(), info))
    if info[0].startswith('м.'):
        info.pop(0)
    
    relocate_will, travel_will = info[0], info[len(info)-1]
    relocate_will = not relocate_will.startswith('не')
    travel_will = not travel_will.startswith('не') or travel_will.endswith(')')
    
    return (relocate_will, travel_will)


def get_relocate_will(info: str) -> bool:
    """Возвращает желание соискателя к переезду
    из строки с информацией о городах, переездах, командировок.
    """
    return get_relocate_and_travel_will(info)[0]


def get_travel_will(info: str) -> bool:
    """Возвращает желание соискателя к командировкам
    из строки с информацией о городах, переездах, командировок.
    """
    return get_relocate_and_travel_will(info)[1]


hh_data['Город'] = hh_data['Город, переезд, командировки'] \
    .apply(get_city_category).astype('category')
display(hh_data['Город'].unique())

hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'] \
    .apply(get_relocate_will)
display(hh_data['Готовность к переезду'].describe())

hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'] \
    .apply(get_travel_will)
display(hh_data['Готовность к командировкам'].describe())

spb_percentage = len(hh_data[hh_data['Город'] == 'Санкт-Петербург']) / len(hh_data) * 100
display(f'Процент соискателей из Санкт-Петербурга: {spb_percentage:.0f}%')

relocate_and_travel_percentage = len(hh_data[hh_data['Готовность к переезду'] & hh_data['Готовность к командировкам']]) / len(hh_data) * 100
display(f'Процент соискателей, готовых одновременно и к переездам, и к командировкам: {relocate_and_travel_percentage:.0f}%')

hh_data.drop('Город, переезд, командировки', axis='columns', inplace=True)
```

5) Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:

<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_unique_subvalues(series: pd.Series) -> list:
    """Возвращает список уникальных значений признака, в котором эти значения
    перечислены через запятую.
    """
    unique_subvalues = set()
    for value in series.to_list():
        for subvalue in value.split(','):
            unique_subvalues.add(subvalue.strip())
    return list(unique_subvalues)


columns_to_split = ['Занятость', 'График']
for column in columns_to_split:
    for subvalue in get_unique_subvalues(hh_data[column]):
        hh_data[subvalue] = hh_data[column].apply(lambda value: subvalue in value)

display(f'{len(hh_data[hh_data["проектная работа"] & hh_data["волонтерство"]])} людей ищут проектную работу и волонтерство')
display(f'{len(hh_data[hh_data["вахтовый метод"] & hh_data["гибкий график"]])} человек хочет работать вахтовым методом и с гибким графиком')

hh_data.drop(columns_to_split, axis='columns', inplace=True)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
exchange_data = pd.read_csv('./data/ExchangeRates.csv')

hh_data['Дата обновления'] = pd.to_datetime(
    hh_data['Обновление резюме'],
    format='%d.%m.%Y %H:%M'
).dt.date

exchange_data['Дата'] = pd.to_datetime(
    exchange_data['date'],
    format='%d/%m/%y'
).dt.date

hh_data['Сумма ЗП'] = hh_data['ЗП'].apply(lambda s: int(s.split()[0]))

currency_to_ISO_map = {
    'грн': 'UAH',
    'USD': 'USD',
    'EUR': 'EUR',
    'белруб': 'BYN',
    'KGS': 'KGS',
    'сум': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'
}
hh_data['Наименование валюты'] = hh_data['ЗП'] \
    .apply(lambda s: currency_to_ISO_map.get(s.split()[1].replace('.', ''), 'руб'))

hh_data = hh_data.merge(
    right=exchange_data,
    how='left',
    left_on=['Дата обновления', 'Наименование валюты'],
    right_on=['Дата', 'currency']
)
hh_data.loc[hh_data['Наименование валюты'] == 'руб', 'close'] = 1

hh_data['ЗП (руб)'] = hh_data['Сумма ЗП'] \
                    * hh_data['close'] \
                    / hh_data['proportion'].fillna(1)

hh_data.drop([
        'ЗП', 'Дата обновления', 'Дата', 'Сумма ЗП', 'Наименование валюты',
        'currency', 'per', 'date', 'time', 'close', 'vol', 'proportion'
    ],
    axis='columns', inplace=True
)
exchange_data.drop('Дата', axis='columns', inplace=True)

display(f'Желаемая медианная ЗП соискателей: {hh_data["ЗП (руб)"].median()/1000:.0f} тыс.')
```

```python
hh_df = hh_data.loc[:, ['ЗП (руб)', 'Готовность к переезду', 'Готовность к командировкам']]

# Формирование вспомогательных датафреймов
relocate_df = hh_df.groupby('Готовность к переезду')['ЗП (руб)'].agg('median')
travel_df = hh_df.groupby('Готовность к командировкам')['ЗП (руб)'].agg('median')

relocate_and_travel_df = hh_df
relocate_and_travel_df['Готовность к переезду и командировкам'] = \
    hh_df['Готовность к переезду'] & hh_df['Готовность к командировкам']
relocate_and_travel_df = relocate_and_travel_df \
    .groupby('Готовность к переезду и командировкам')['ЗП (руб)'].agg('median')

relocate_or_travel_df = hh_df
relocate_or_travel_df['Готовность к переезду или командировкам'] = \
    hh_df['Готовность к переезду'] | hh_df['Готовность к командировкам']
relocate_or_travel_df = relocate_or_travel_df \
    .groupby('Готовность к переезду или командировкам')['ЗП (руб)'].agg('median')

# Датафрейм для построения диаграммы
hh_df = pd.DataFrame({
    'Переезд': relocate_df,
    'Командировки': travel_df,
    'И то, и другое': relocate_and_travel_df,
    'Или то, или другое': relocate_or_travel_df
}).transpose()

# Столбчатая диаграмма
fig = px.bar(
    hh_df,
    x=hh_df.index,
    y=[hh_df.columns[0], hh_df.columns[1]],
    color_discrete_sequence=('#ca4e4e', '#4eca5c'),
    opacity=.85,
    text_auto='.3s'
)

# Настройка отображения
fig.update_layout(
    title=dict(
        text='Зависимость медианной заработной платы<br>от признаков «Готовность к переезду» и «Готовность к командировкам»',
        x=.5, xanchor='center'
    ),
    width=750, height=500,
    margin=dict(l=100, r=60, t=100, b=70),
    legend=dict(
        title='Значение признака:',
        x=.5, xanchor='center', orientation='h'
    )
)
fig.update_xaxes(title='')
fig.update_yaxes(title='Медианная заработная плата (руб.)')

fig.show('svg')
fig.write_html('./plotly/6.html')
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
log_age = np.log(hh_data['Возраст'])

# Гистограмма
fig = px.histogram(
    log_age,
    nbins=90
)

# Расчитывание значений метода z-отклонений
mu = log_age.mean()
sigma = log_age.std()
lower_bound = mu - 3*sigma
upper_bound = mu + 3*sigma

# Линия среднего распределения
fig.add_vline(
    mu,
    line=dict(width=3, color='Black'),
    opacity=.85
)
# Линия нижней границы интервала 3-х сигм
fig.add_vline(
    lower_bound,
    line=dict(width=3, dash='dash', color='Black'),
    opacity=.85
)
# Линия верхней границы интервала 3-х сигм
fig.add_vline(
    upper_bound,
    line=dict(width=3, dash='dash', color='Black'),
    opacity=.85
)

# Настройка отображения
fig.update_layout(
    title=dict(
        text='Распределение признака "Возраст" в логарифмическом масштабе',
        x=.5, xanchor='center'
    ),
    width=750, height=500,
    margin=dict(l=90, r=60, t=80, b=70),
    showlegend=False
)

fig.show('svg')
fig.write_html('./plotly/log_age.html')

# Избавление от выбросов с помощью метода z-отклонений
upper_bound = mu + 4*sigma  # Послабление на одну сигму в правую сторону
outliers = hh_data.loc[(log_age < lower_bound) | (log_age > upper_bound), :]
display(outliers)

hh_data.drop(outliers.index, axis='index', inplace=True)
```

* Логарифмическое распределение признака "Возраст" ассиметрично и смещено в правую сторону.
* Под категорию выбросов попали соискатели с возрастом **15** и **100** лет.
